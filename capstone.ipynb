## Complaints type

```python
# how many complaint types


df_311['complaint_type'].unique().shape

df_311['complaint_type'] = df_311['complaint_type'].str.lower()

df_311['complaint_type'].unique()
```

```python
sum(df_311['complaint_type'] == 'elevator')
sum(df_311['complaint_type'] == 'electric')
```

```python
for x in df_311['complaint_type'].unique():
    xtrue = sum(df_311['complaint_type'] == x)
    if  xtrue >= 80000:
        print('Type: %s has %i cases' % (x, xtrue))
```
